## Encapsulación
- Ejemplo son las clases agrupan atributos y metodos

```java
// Ejemplo de definición de una clase
public class Persona {
    // Atributos
    private String nombre;
    private int edad;
    
    public Persona (String nombrePersona, int edadPersona) {
        nombre = nombrePersona;
        edad   = edadPersona;
    }
    
    // Metodos
    public void presentarse () {
        System.out.println ("Soy " + nombre);
        System.out.println ("Edad es :" + edad);
        
    }
    public void caminar() {
        System.out.println ("Estoy caminando...");
    }
}
```

```java
// Ejemplo de creación de objetos
Persona p = new Persona ("Maria", 33);
p.presentarse ();
p.caminar ();
```

## Herencia
- Concepto para modelar los problemas de forma más natural
  - Persona
  - Estudiante que es Persona
  - Trabajador que es Persona
  - EstudianteColegio que es Estudiante y es Persona

<img src="UML-Persona.png" width="300">

- Permite reutilizar código:
  - No volver a escribir lo que estaba escrito:
  - No volver a codificar lo que ya está codificado en otra parte
  - Evitar errores al escribir nuevo código
  
- Elementos:
  - Clase base o superclase
  - Clases hijas o subclases

```java
// Ejemplo Persona, Estudiante, Trabajador
public class Persona {
    // Atributos
    String nombre;
    int edad;
    
    public Persona () {
        nombre = "Desconocido";
        edad   = 0;
    }
    
    public Persona (String nombre, int edad) {
        this.nombre = nombre;
        this.edad   = edad;
    }
    
    // Metodos
    public void presentarse () {
        System.out.println ("Soy " + nombre);
        System.out.println ("Edad es :" + edad);
        
    }
    
    public void caminar() {
        System.out.println ("Estoy caminando...");
    }
}
```

<img src="UML-Persona.png" width="400">

```java
Persona p = new Persona ("Maria", 33);
p.presentarse ();
```

```java
// Subclase Estudiante
public class Estudiante extends Persona {
    int codigo;
    public Estudiante () {
        super ();
        codigo = 0;            
    }
    
    public Estudiante (String nombre, int edad, int codigo) {
        super (nombre, edad);
        this.codigo = codigo;       
    }
    
    public void presentarse () {
        super.presentarse ();
        System.out.println ("Codigo es :" + codigo);  
    }
    
    public void estudiar () {
        System.out.println ("Estoy estudiando...");
    }
} 
```

```java
// Objeto con constructor por defecto
Estudiante eX = new Estudiante ();
eX.presentarse ();
```

```java
// Objetos
Estudiante e1 = new Estudiante ("Maria", 22, 1010);
e1.presentarse ();
e1.caminar ();
e1.estudiar ();
```

```java
// Subclase EstudianteColegio
public class EstudianteColegio extends Estudiante {
    int grado;
    
    public EstudianteColegio (String nombre, int edad, int codigo, int grado)  {
        super (nombre, edad, codigo);
        this.grado = grado;
    }
    
    public void presentarPruebasSaber11 () {
        System.out.println ("Presentando las pruebas Saber 11...");  
    }
}
```

```java
// Objetos
EstudianteColegio ec = new EstudianteColegio ("Ana", 12, 3030, 11);
ec.presentarse (); 
```

```java
ec.presentarPruebasSaber11();
```

```java
ec.caminar ();
```

### Ejemplo de Animales
<img src="uml-animales.png" width="400">

```java
public abstract class Animal {
    public abstract void hacerRuido ();
}

public class Perro extends Animal {
    public void hacerRuido () {
        System.out.println ("guau guau");
    }
    
    public void hacerRuido (String estado) {
        if (estado == "contento")
            System.out.println ("Mueve la cola y dice guau guau");
        else if (estado == "bravo")
            System.out.println ("Muerde y dice GUAU GUAU");
        else
            hacerRuido ();      
    }
}

public class Gato extends Animal {
    public void hacerRuido () {
        System.out.println ("miau miau");
    }    
}

public class Vaca extends Animal {
    public void hacerRuido () {
        System.out.println ("muu muu");
    }    
}
public class Tigre extends Animal {
    public void hacerRuido () {
        System.out.println ("grurrr grurrr");
    }    
}
```

```java
Perro firulais = new Perro ();
firulais.hacerRuido ();
firulais.hacerRuido ("bravo");
```

```java
Gato garfiel = new Gato ();
garfiel.hacerRuido ();
```

```java
Vaca lola = new Vaca ();
lola.hacerRuido();
```

## Ejemplo zoologico:
- Multiples animales
- Diferentes tipos
- Ingreso en una pieza y necesitas saber cual es por su tipo de ruido;

```java
Perro firulais = new Perro ();
Gato garfiel = new Gato ();
Vaca lola = new Vaca ();
Tigre bengala = new Tigre ();
Perro p = new Perro ();
Vaca  c = new Vaca ();
Tigre t = new Tigre ();

List <Animal> zoologico = new ArrayList <>();
zoologico.add (firulais);
zoologico.add (lola);
zoologico.add (bengala);
zoologico.add (c);
zoologico.add (p);
zoologico.add (t);

for (Animal x:zoologico) {
    x.hacerRuido ();
}
```
