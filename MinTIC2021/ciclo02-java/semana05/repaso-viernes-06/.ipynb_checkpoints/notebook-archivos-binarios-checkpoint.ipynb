## Excepciones de formato

```java
// Se atrapa una excepcion de formato cuando se lee una cadena en vez de un número
import java.util.Scanner;
//import java.lang.InputMismatchException;

Scanner lector = new Scanner (System.in);

long x;
try {
    x = lector.nextLong ();
}catch (InputMismatchException ex) {
    System.out.println ("Inserte un valor númerico.");
    x=5;
}

System.out.println ("Valor de x: " + x);
```

## Archivos Binarios
- Archivos Texto con format:
  - FileWriter/FileReader
  - PrintWriter/BufferedReader
  
- Archivos Binarios:
  - **FileInputStream / FileOutputStream** : para manejo en dispositivos de archivos
  - **ObjectInputStream / ObjectOutputStream** : para crear flujos binarios
  
- Los objetos que se vayan a serializar (escribir o leer) deben implementar la interfaz "serializable"

```java
class Estudiante implements Serializable {
//...

}
```

```java
// Escritura 
String nombres = "Ana Maria";
String  correo  = "ana@gmail.com";
long    numero  = 11;

// Crear el flujo para escribir en archivos
FileOutputStream fos   = new FileOutputStream ("datos.dat");
ObjectOutputStream oos = new ObjectOutputStream (fos);

// Escritura
oos.writeObject (nombres);
oos.writeObject (correo);
oos.writeObject (numero);
```

```java
// Lectura
String nombres;
String  correo;
long    numero;

System.out.println (nombres + ", "+correo+", "+numero);

FileInputStream fis = new FileInputStream ("datos.dat");
ObjectInputStream ois = new ObjectInputStream (fis);

nombres = (String) ois.readObject();
correo = (String) ois.readObject();
numero = (long) ois.readObject();

System.out.println (nombres + ", "+correo+", "+numero);
```

```java
// Escritura de todo un objeto estudiante
public class Estudiante implements Serializable {
    // Atributos
    String nombres;
    String correoInstitucional;
    long numeroCelular;
        
    // Constructor
    public Estudiante () {
        this.nombres = "";
        this.correoInstitucional = "";
        this.numeroCelular = 0;
    }
    
    public Estudiante (String nombres, String correoInst, long nroCelular){
        this.nombres = nombres;
        this.correoInstitucional = correoInst;
        this.numeroCelular = nroCelular;
    }
    
    // Metodos
    public void presentarse () {
        System.out.println ("NOMBRES: " + nombres + " CORREO: " + correoInstitucional + " CELULAR" + numeroCelular);
    }            
}
```

```java
// Escritura de todo el estudiante como objeto
Estudiante e = new Estudiante ("ana", "aa", 11);
// Crear el flujo para escribir en archivos
FileOutputStream fos   = new FileOutputStream ("datos-estudiante.dat");
ObjectOutputStream oos = new ObjectOutputStream (fos);

oos.writeObject (20);
oos.writeObject (e);
```

```java
// Lectura de todo un objeto y conversion a estudiante
FileInputStream fis = new FileInputStream ("datos-estudiante.dat");
ObjectInputStream ois = new ObjectInputStream (fis);

int size     = (int) ois.readObject(); 
Estudiante e = (Estudiante) ois.readObject();
System.out.println (size);
e.presentarse ();
```
