## Conversiones entre tipos de clases : Casting, Casteo
- Conversion entre diferentes tipos
- Para poder convertir entre tipos, estos deben "compatibles"

```java
// Casting de numeros
int a = 10;
long b = 20;  // Este tipo es como un entero más grande

a = b;
```

```java
// Conversion del long a entero : casting
a = (int)b;
System.out.println (a);
```

```java
// Conversion de decimal a entero
int x;+
double y = 10.56;

x = y;
```

```java
// Conversion de decimal a entero
int x;
double y = 10.56;

x = (int)y;
System.out.println (x);
```

```java
// Objeto "o" contiene una cadena pero no es de Clase cadena

Object o = new String ("Hola Mundo");

// Recupero esa cadena del objeto o
String s = (String) o;

// La uso como cadena
System.out.println (s);
s.contains("a")
```

## Clases para archivos binarios serializable
- Se escriben en y leen en formato binario objetos completos:
- Clases para la Lectura:
  - **FileInputStream** : Dispositivos de lectura desde archivos
  - **ObjectInputStream** : Flujo de datos binario para leer
 
- Clases para la Escritura:
  - **FileOutputStream** : Dispositivos de escritura en archivos
  - **ObjectOutputStream** : Flujo de datos binario para enviar
  
- Los clases de objetos que se van a escribir/leer deben ser serializable, es decir implementar la interface Serialiable

```java
// Ejemplo de clase que implementa la interface Serializable
class Estudiante implements Serializable {
//..
//..
}
```

## Ejemplos Archivos binarios serializable

```java
// Escritura
String nombre   = "Luis";
String correo   = "luis@gmail.com";
long    numero  = 11;

// Creación archivo para escribir flujo de datos binario
FileOutputStream fos = new FileOutputStream ("estudiantes-binario.dat");
ObjectOutputStream oos = new ObjectOutputStream (fos);

// Escritura con metodo writeObjetc
oos.writeObject (nombre);
oos.writeObject (correo);
oos.writeObject (numero);

// Cierro el flujo
oos.close ();
```

```java
// Lectura
String nombre;
String correo;
long    numero;

System.out.println ("Valores iniciales: ");
System.out.println (nombre + ", " + correo + "," + numero);

// Creación archivo para leer desde un flujo de datos binarios
FileInputStream fis = new FileInputStream ("estudiantes-binario.dat");
ObjectInputStream ois = new ObjectInputStream (fis);

// Lectura on metodo readObject
nombre = (String )ois.readObject();
correo = (String )ois.readObject();
numero = (long )ois.readObject();

System.out.println ("Valores recuperados: ");
System.out.println (nombre + ", " + correo + ", " + numero);
```

## Escritura/Lectura en binario de Objetos de Clase Estudiante

```java
public class Estudiante implements Serializable {
    // Atributos
    String nombres;
    String correoInstitucional;
    long numeroCelular;
        
    // Constructor
    public Estudiante () {
        this.nombres = "";
        this.correoInstitucional = "";
        this.numeroCelular = 0;
    }
    
    public Estudiante (String nombres, String correoInst, long nroCelular){
        this.nombres = nombres;
        this.correoInstitucional = correoInst;
        this.numeroCelular = nroCelular;
    }
    
    // Metodos
    public void presentarse () {
        System.out.println ("NOMBRES: " + nombres + " CORREO: " + correoInstitucional + " CELULAR" + numeroCelular);
    }            
}
```

```java
// Escritura de Estudiante
Estudiante e1 = new Estudiante ("ana", "ana@gmail.com", 1234);

// Creación archivo para escribir flujo de datos binario
FileOutputStream fos = new FileOutputStream ("estudiante.dat");
ObjectOutputStream oos = new ObjectOutputStream (fos);

// Escritura con metodo writeObject
oos.writeObject(e1);

// Cierro el flujo
oos.close ();
```

```java
// Lectura de Estudiante
// Creación archivo para leer desde un flujo de datos binarios
FileInputStream fis = new FileInputStream ("estudiante.dat");
ObjectInputStream ois = new ObjectInputStream (fis);

// Lectura con metodo readObject y la conversion (casting)
Estudiante e = (Estudiante)ois.readObject ();
e.presentarse ();
```

## Implementación serializable en el ejemplo de estudiantes
- Solo se modifica el DAO
  - Se crea un metodo guardarse() que se ejecuta cada que se modifica la lista del DAO
      - El metodo guardarse() escribe siempre toda la lista desde el inicio al final, es decir, el archivo siempre está actualizado.
      - Se guarda primero el numero de elementos y después se recorre la lista guardando cada estudiante
  - Se crea un método recuperarse que se ejecuta cuando en el metodo crear del DAO, es decir, cuando el controlador crea el DAO para iniciar el programa.
    - El metodo recuperarse abre el archivo y recupera primero el numero de elementos y despues en un ciclo recupera estudiante por estudiante y lo agrega a la lista del DAO.
