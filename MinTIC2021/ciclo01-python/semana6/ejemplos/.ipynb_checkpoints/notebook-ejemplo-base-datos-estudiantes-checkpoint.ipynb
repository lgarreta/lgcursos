- Todo se va a manejar en lista: listas de listas

- Eliminar

- Imprimir
- Cambiar nota

```python
# Ejemplo programa con menu
## FUNCIONES
# Adiciona un nuevo estudiante a la lista
def adicionar (listaEstudiantes):
    nombre = input ("NOMBRE: ")
    codigo = input ("CODIGO: ")
    nota   = float (input ("NOTA: "))    
    
    estudiante = [nombre, codigo, nota]
    
    listaEstudiantes.append (estudiante)
    
def borrar ():
    print ("Aun no implementada")
    
def buscar ():
    print ("Aun no implementada")

def imprimir (listaEstudiantes):
    # Abrir el archivo
    N = len (listaEstudiantes)
    i = 0
    while (i < N):
        estudiante = listaEstudiantes [i]
        nombre = estudiante [0]
        codigo = estudiante [1]
        nota   = estudiante [2]
        print (nombre)           # escribir en el archivo
        print (codigo)           # 
        print (nota)
        
        i+=1
    #
    # Cierrar el archivo
#

def guardarArchivo (listaEstudiantes):
    archivo = open ("agenda.txt", "w") # Abrir el archivo
    
    N = len (listaEstudiantes)
    i = 0
    while (i < N):
        estudiante = listaEstudiantes [i]
        nombre = estudiante [0]
        codigo = estudiante [1]
        nota   = str (estudiante [2])
        archivo.write (nombre) #print (nombre)           # escribir en el archivo
        archivo.write ("\n")
        archivo.write (codigo) #print (codigo)           # 
        archivo.write ("\n")
        archivo.write (nota)   #print (nota)
        archivo.write ("\n")
        i+=1
    #
    archivo.close () # Cierrar el archivo
    
## VARIABLES
# Estudiante : [Nombre, codigo, nota]
listaEstudiantes = [] # Ej. [ ["juan", 1010, 4.5], ["maria", 2020, 3.5], ["andrea", 3030, 2.8]]

## PRINCIPAL    
while (True):
    print (">>>>>>>>>>> OPCIONES <<<<<<<<<<<<<<<<<<")
    print ("0. Salir")
    print ("1. Adicionar estudiante")
    print ("2. Borrar estudiante")
    print ("3. Buscar si existe estudiante")
    print ("4. Guardar en archivo.")
    print ("5. Imprimir")
    print ()
    opcion = int (input ("Ingrese la opción"))
    print (">>>>>>>>>>>>>>>>>><<<<<<<<<<<<<<<<<<<<<<")
    print ()
       
    if (opcion==0):
        break
    elif (opcion==1):
        adicionar (listaEstudiantes)
    elif (opcion==2):
        borrar()
    elif (opcion==3):
        buscar ()
    elif (opcion==5):
        imprimir (listaEstudiantes)
    elif (opcion==4):
        guardarArchivo (listaEstudiantes)
    else:
        print ("Opción no valida")
    #
    print ()
#
print ()
print ("Adios")
    
    
```
